## 1.4 Buatlah status BMI untuk setiap siswa.

Underwieght : < 18.5

Normal : 18.54 - 24.9

Overweight : 25.0 - 29.9

Obesity : 30.0 - 34.9



```python
#Status BMI
def kategori_bmi(bmi):
    if bmi < 18.5:
      return 'Underweight'
    elif 18.5 <= bmi < 24.9:
      return "Normal"
    elif 25.0 <= bmi < 29.9:
      return  "Overweight"
    elif 30.0 <= bmi < 34.9:
      return "Obesity"
    else:
        return 'Over'

data['Kategori BMI'] = data['BMI'].apply(kategori_bmi)

display(data[['Nama', 'BMI', 'Kategori BMI']])
```

Kode diatas memberikan pelabelan terhadap varibel BMI, dimana ketika nilai BMI pada dataframe memenuhi suatu kondisi tertentu maka ia dilabeli susuai dengan kondisi yant terpenuhi.

# Soal No. 2
Perusahaan investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

```python
#NO 2
def hitung_investasi(Modal, Waktu):
    if not (0 <= Modal <= 1000000000):
        return "Error: Modal tidak sesuai ketentuan (0-1.000.000.000)."

    if not (1 <= Waktu <= 25):
        return "Error: Jangka waktu tidak sesuai ketentuan (1-25 tahun)."

    # Akumulasi investasi setiap tahun
    for i in range(1, Waktu + 1):
        Bunga = Modal * bunga
        Modal += Bunga
        print(f">> Tahun {i}, total investasi anda : {Modal:.1f}")


Modal = int(input("Modal awal(0 - 1000000000): "))
Waktu = int(input("Waktu(1 - 25 tahun): "))
bunga = 0.25

hitung_investasi(Modal, Waktu)
```

fungsi hitung_investasi adalah fungsi untuk menghitung jumlah tabungan yang diinvestasikan berdasarkan modal awal dan jangka waktu sesuai yang diinputkan pengguna.

Insight: Jumlah tabungan yang didapat pertahun

# Soal No.3

Buatlah program yang dapat digunakan untuk menghasilkan sebanyak n bilangan prima pertama, dengan n merupakan input user. Jelaskan secara singat alur program tersebut

```python
#NO 3
def apakah_prima(bil):
  if bil < 2:
    return False
  for i in range(2, int(bil ** 0.5) + 1):
    if bil % i == 0:
      return False
  return True

def bil_prima_pertama(n):
  bilangan_prima = []
  bil = 2
  while len(bilangan_prima) < n:
    if apakah_prima(bil):
      bilangan_prima.append(bil)
    bil += 1
  return bilangan_prima

n = int(input("Masukkan jumlah bilangan prima yang diinginkan: "))

prima_pertama = bil_prima_pertama(n)
print(f"{n} bilangan prima pertama: {prima_pertama}")
```

**Alur program:**

1. User input bilangan sebagai permintaan untuk menunjukan jumlah bilangan prima yang ingin dimunculkan
2. Sistem menampilkan bilangan prima sebanyak yang diinputkan user. Sebelum itu sistem, apakah bilangan tersebut prima atau bukan. Hal itu dilakukan dalam proses fungsi apakah_prima(bil).

Insight: Bilangan prima sebanyak yang diinputkan user

```python
#NO 4
n = int(input("Masukan angka: "))

# Loop dari 1 hingga n (membuat baris-baris piramida)
for i in range(1, n + 1):
  print(' ' * (n - i) + '*' * (2 * i - 1), end='')#bagian kiri

  print(' ' *(2*(n-i) +1), end = '') # spasi antar piramida

  print('*' * (2 * i - 1)) #bagian kanan
```

Perulangan yang dimana output dari setiap perulangnnya adalah * dan spasi dan membentuk piramida. Ketika masih memenuhi kondisi maka perulangan itu masi akan terus dilakukan

Insight : piramida

```python
def diamond(n):
    # Bagian atas (segitiga sama kaki)
    for i in range(1, n + 1):
        char = chr(65 + (i - 1) % 26)  # Menggunakan nilai ASCII untuk huruf (65 = 'A')
        print(' ' * (n - i) + ' '.join([char] * i))  # Spasi antar huruf pada baris

    # Bagian bawah (segitiga sama kaki terbalik)
    for i in range(n - 1, 0, -1):
        char = chr(65 + (i - 1) % 26)
        print(' ' * (n - i) + ' '.join([char] * i))

# Menentukan tinggi diamond
n = int(input())
diamond(n)
```

def diamond adalah fungsi untuk membentuk diamond dengan dua kali perulangan. Perulangan pertama yaitu membentuk diamond dari atas dengan char yang menggunakan nilai ASCII, selanjutnya perulangan dilanjutkan dari bawah.

Insight : Diamond dengan abjad yang tingginya sesuai dengan input user
